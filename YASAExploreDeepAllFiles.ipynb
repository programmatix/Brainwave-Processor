```python
import os
import pandas as pd
import run_feature_pipeline

errors = []
dataframes = []

for root, dirs, files in os.walk(input_dir):
    for dir_name in dirs:
        try:
            input_file = os.path.join(root, dir_name, "raw.with_features.csv")
            log("Processing file: " + input_file)
            if os.path.exists(input_file):
                df = pd.read_csv(input_file)
                df['filename'] = input_file
                dataframes.append(df)
        except Exception as e:
            log("Error processing file: " + input_file)
            errors.append("Error processing file: " + input_file + " - " + str(e))
            log(e)

# Concatenate all dataframes into a single dataframe
if dataframes:
    combined_df = pd.concat(dataframes, ignore_index=True)
else:
    combined_df = pd.DataFrame()

for err in errors:
    log(err)
df = combined_df
df
```

```python
useful_cols = ['F8-M1_eeg_skew',
               'F8-M1_eeg_abspow',
               'F8-M1_eeg_dt',
               'F8-M1_eeg_kurt',
               'F8-M1_eeg_ds']
```

```python
filenames = list(df_deep['filename'].unique())
```

```python
# Bear in mind we're interested in the next day's feeling.  '07-15' means the night of 07-15, and how I was feeling on 07-16.
good_days = {
    '2024-07-11': 'good', # one note
    '2024-07-12': 'tired', # just on enote
    '2024-07-13': 'n/a',
    '2024-07-14': 'n/a',
    '2024-07-15': 'ok',
    '2024-07-16': 'good',
    '2024-07-17': 'tired', # very sure, lots of notes
    '2024-07-21': 'tired/good mix',
    '2024-07-22': 'maybe good',
    '2024-09-01': 'tired', # very sure
    '2024-09-02': 'tired', # very sure
    '2024-09-03': 'tired', # very sure
}
```

```python
import pandas as pd

# Initialize an empty DataFrame with filenames as the index
df_stats = pd.DataFrame(index=filenames)
pd.options.display.float_format = '{:.2f}'.format

# Iterate over each filename
for filename in filenames:
    date_str = '-'.join(filename.split('\\')[-2].split('-')[0:3])
    df_stats.loc[filename, 'date'] = date_str
    # Iterate over each column in useful_cols
    for col in useful_cols:
        # Calculate the statistics for each column and store them in df_stats
        values = df_deep[df_deep['filename'] == filename][col]
        df_stats.loc[filename, 'N3 count'] = (values.count() * 30) / 60
        df_stats.loc[filename, col + '_mean'] = values.mean()
        #df_stats.loc[filename, col + '_min'] = values.min()
        #df_stats.loc[filename, col + '_max'] = values.max()
        #df_stats.loc[filename, col + '_25%'] = values.quantile(0.25)
        #df_stats.loc[filename, col + '_75%'] = values.quantile(0.75)
        try:
            
            df_stats.loc[filename, 'feeling_next_day'] = good_days.get(date_str, 'unknown')
        except Exception as e:
            df_stats.loc[filename, 'feeling_next_day'] = str(e)
            pass

#df_stats.reset_index(inplace=True)
df_stats
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Filter the DataFrame for 'good' and 'tired' feelings
filtered_df = df_stats[df_stats['feeling_next_day'].isin(['good', 'tired'])]

# Select relevant features for visualization
features = ['N3 count', 'F8-M1_eeg_skew_mean', 'F8-M1_eeg_abspow_mean', 'F8-M1_eeg_dt_mean', 'F8-M1_eeg_kurt_mean', 'F8-M1_eeg_ds_mean']

# Create a pairplot to visualize the distribution of features
sns.pairplot(filtered_df, vars=features, hue='feeling_next_day', palette={'good': 'green', 'tired': 'red'})
plt.suptitle('Feature Distribution for Good and Tired Days', y=1.02)
plt.show()
```
