# Find alignment between YASA and my scoring
First we align the stages so they are comparable.

```python
aligned = joined.copy()

def map_yasa_stage(stage):
    if stage == 'W':
        return 'Wake'
    elif stage == 'N3':
        return 'Deep'
    else:
        return 'Non-Deep'

aligned['YasaStageAligned'] = aligned['Stage'].apply(map_yasa_stage)

def map_manual_stage(stage):
    if stage == 'Ambiguous Deep':
        return 'Deep'
    else:
        return stage

aligned['ManualStageAligned'] = aligned['ManualStage'].apply(map_manual_stage)

desired_order = ['Stage', 'YasaStageAligned', 'ManualStage', 'ManualStageAligned']
remaining_cols = [col for col in aligned.columns if col not in desired_order]
new_order = desired_order + remaining_cols
aligned = aligned[new_order]
```

```python
confusion_matrix = pd.crosstab(aligned['YasaStageAligned'], aligned['ManualStageAligned'], rownames=['YasaStageAligned'], colnames=['ManualStageAligned'])
confusion_matrix
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.clf()

# Assuming `confusion_matrix` is already created
plt.figure(figsize=(10, 7))
sns.heatmap(confusion_matrix, annot=True, fmt='d', cmap='Blues')

plt.title('Confusion Matrix between Stage and ManualStage')
plt.xlabel('ManualStage')
plt.ylabel('YasaStage')
plt.show()
```
