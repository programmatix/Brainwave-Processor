```python
df_deep = df[df['Stage'] == 'N3']

numerical_columns = df_deep.select_dtypes(include=['number']).columns
columns_to_remove = [col for col in numerical_columns if "p2" in col or "c7" in col or "Confidence" in col or "Epoch" in col or "StageInt" in col]
numerical_columns = [col for col in numerical_columns if col not in columns_to_remove]

# Create new DataFrames with only numerical columns
df_deep_numerical = df_deep[numerical_columns]
```

```python
import pandas as pd

# Calculate the mean of each feature for df1_deep and df2_deep
df_deep_means = df_deep_numerical.mean()

# Create a DataFrame to store the comparison results
comparison_df = pd.DataFrame({
    'Feature': df_deep_means.index,
    'df_deep_mean': df_deep_means.values,
})

# Determine which features were generally higher or lower
comparison_df['Difference'] = comparison_df['df1_deep_mean'] - comparison_df['df2_deep_mean']
comparison_df['DifferencePct'] = (comparison_df['Difference'] / comparison_df['df2_deep_mean']) * 100
comparison_df['DifferencePctAbs'] = comparison_df['DifferencePct'].abs()
comparison_df['Higher_in'] = comparison_df['Difference'].apply(lambda x: 'df1_deep' if x > 0 else 'df2_deep')

pd.set_option('display.max_rows', None)

# Display the comparison results
sorted = comparison_df.sort_values(by='DifferencePctAbs', ascending=False)
top_features = sorted[sorted['DifferencePctAbs'] > 10]
top_features
```

```python
frontal_top_features = top_features[top_features['Feature'].str.contains('F')]
frontal_top_features
```

```python
import pandas as pd

# Assuming top_features and df1 are already defined
top_feature_names = top_features['Feature'].tolist()

# Define a threshold for high and low values (e.g., mean ± 1 standard deviation)
thresholds = {}
for feature in top_feature_names:
    mean_value = df1_deep[feature].mean()
    std_value = df1_deep[feature].std()
    thresholds[feature] = {
        'high': mean_value + std_value,
        'low': mean_value - std_value
    }

# Filter df1 to get the epochs where top features are particularly high or low
high_low_epochs = {}
for feature in top_feature_names:
    high_epochs = df1_deep[df1_deep[feature] > thresholds[feature]['high']]['Epoch']
    low_epochs = df1_deep[df1_deep[feature] < thresholds[feature]['low']]['Epoch']
    high_low_epochs[feature] = {
        'high': high_epochs.tolist(),
        'low': low_epochs.tolist()
    }

# Display the high and low epochs for each top feature
for feature, epochs in high_low_epochs.items():
    print(f"Feature: {feature}")
    print(f"High epochs: {epochs['high']}")
    print(f"Low epochs: {epochs['low']}")
    print()
```
