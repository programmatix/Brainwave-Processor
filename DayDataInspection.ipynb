```python
import os
import pandas as pd

input_dir = "C:\\dev\\play\\brainwave-data-day"
stats_df = pd.read_csv(input_dir + os.path.sep + "day_stats.csv")

```

```python
from datetime import datetime
import os

dfs = []

for root, dirs, files in os.walk(input_dir):
    for idx, dir_name in enumerate(dirs):
        input_file = os.path.join(root, dir_name, "raw.output.csv")
        if os.path.exists(input_file):
            df = pd.read_csv(input_file)
            df['TimestampUK'] = pd.to_datetime(df['TimestampUK'])
            day_and_night_of_date = datetime.strptime(dir_name, "%Y-%m-%d-%H-%M-%S")
            day_and_night_of = day_and_night_of_date.date().isoformat()
            df['dayAndNightOf'] = day_and_night_of
            dfs.append(df)

eeg_df = pd.concat(dfs, ignore_index=True)
```
